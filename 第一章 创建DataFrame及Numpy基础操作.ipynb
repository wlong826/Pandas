### (6) æŸ¥æ‰¾ç¬¬ä¸€åˆ—çš„å±€éƒ¨æœ€å¤§å€¼ä½ç½®ï¼šNumpy.sign()

```python
#æ€è·¯ï¼šæŸ¥æ‰¾æ¯”å®ƒå‰ä¸€ä¸ªå’Œåä¸€ä¸ªæ•°å­—éƒ½å¤§æ•°å­—
tem = np.diff(np.sign(np.diff(df['score'])))
np.where(tem == -2)[0] + 1
```

### (8) è®¡ç®—ä¸¤åˆ—æ¬§å¼è·ç¦»ï¼šNumpy.linalg.norm()
