```python
import pandas as pd
import numpy as np
```

```python
data = {"grammer":['Python', 'C', 'Java', 'GO', 'css', 'SQL', 'PHP', 'Python'],
       "score":[1.0, 2.0, 6.0, 4.0, 5.0, 6.0, 7.0, 10.0]}
df = pd.DataFrame(data)
df
```

## 1 查看前(后)5行数据

```python
#前5行：
df.head() # 默认是5
```

```python
#后5行：
# 方法一：
df.tail() # 默认是5
# 方法二：
df.iloc[-5:, :]
```

## 2. 查看所有列名

```python
df.columns
```

## 3. 查看数据行列数 

```python
df.shape
```

## 4. 查看行数

```python
#方法一：
df.shape[0]
#方法二；
len(df)
```

## 5. 查看列数

```python
#方法一：
len(df.columns)
#方法二：
df.shape[1]
```

## 6. 查看索引、数据类型和内存信息

```python
df.info()
```

## 7. 查看索引

```python
df.index
```

## 8. 查看数值型列的汇总统计

```python
df.describe()
```

## 9. 查看每列的数据类型

```python
df.dtypes
```

## 10. 查看某列共有几种元素

```python
#方法一：
df['grammer'].nunique()
#方法二：
len(df['grammer'].unique())
#方法三：
len(df.groupby('grammer'))
#方法四：
len(df['grammer'].value_counts())
```

## 11. 查看pandas与numpy版本

```python
print(pd.__version__)
print(np.__version__)
```

## 12. 查看某一列各元素出现的次数

```python
df['grammer'].value_counts()
```

## 13. 查看某列每个字符串的长度

```python
# 方法一：
df['grammer'].map(lambda x: len(x))
# 方法二：
df['grammer'].str.len()
```
